In this notebook, first project of the [Deep Reinforcement Learning Nanodegree](https://www.udacity.com/course/deep-reinforcement-learning-nanodegree--nd893) is shown.

```python
from unityagents import UnityEnvironment
import numpy as np
```

```python
from agent import Agent
from train_agent import dqn
```

```python
env = UnityEnvironment(file_name="envs/Banana.app")
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
env_info = env.reset(train_mode=True)[brain_name]

action_size = brain.vector_action_space_size

state = env_info.vector_observations[0]
state_shape = state.shape

agent = Agent(state_shape=state_shape, action_size=action_size, seed=0)

print("Start training...")
scores = dqn(env, agent, brain_name, n_episodes=500)
```

```python
import matplotlib.pyplot as plt


# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```
